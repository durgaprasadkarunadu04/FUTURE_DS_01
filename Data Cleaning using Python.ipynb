```python
import pandas as pd
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')
```

```python
print("Dataset loaded. Shape:", df.shape)
```

```python
required_columns = [
    'Order ID', 'Order Date', 'Ship Date', 'Ship Mode',
    'Customer Name', 'Segment', 'Region', 'State', 'City',
    'Category', 'Sub-Category', 'Product Name',
    'Sales', 'Quantity', 'Discount', 'Profit'
]
df = df[required_columns]
```

```python
df = df.drop_duplicates()
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month-Year'] = df['Order Date'].dt.to_period('M').astype(str)
```

```python
print("Cleaning done. Final Shape:", df.shape)
df.head()
```

```python
print("Total Sales: $", round(df['Sales'].sum(), 2))
print("Total Profit: $", round(df['Profit'].sum(), 2))
print("Total Orders:", df['Order ID'].nunique())

print("\n--- Sales by Category ---")
print(df.groupby('Category')['Sales'].sum().sort_values(ascending=False).round(2))

print("\n--- Profit by Category ---")
print(df.groupby('Category')['Profit'].sum().sort_values(ascending=False).round(2))

print("\n--- Sales by Region ---")
print(df.groupby('Region')['Sales'].sum().sort_values(ascending=False).round(2))

print("\n--- Sales by Segment ---")
print(df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).round(2))

print("\n--- Top 10 Products by Sales ---")
print(df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).round(2))
```

```python
df.to_csv('cleaned_superstore.csv', index=False)
print("â cleaned_superstore.csv saved successfully!")
```
